```python
import random
import string

# Function to generate a large random string block (~100 KB)
def generate_random_string(size=100000):
    return ''.join(random.choices(string.ascii_letters + string.digits + ' ', k=size))

# Function to create large file of given size (in MB)
def create_large_file(filename, size_in_mb):
    size_bytes = size_in_mb * 1024 * 1024  # MB to bytes
    block = generate_random_string()  # ~100KB per block
    block_size = len(block)

    with open(filename, 'w') as f:
        for _ in range(size_bytes // block_size):
            f.write(block)
    print(f"{filename} of size ~{size_in_mb}MB created.")
```

```python
# Create a 10MB file (safe to test in Colab or PC)
create_large_file("file_10MB.txt", 10)

# You can try 50MB or 100MB too:
# create_large_file("file_50MB.txt", 50)
# create_large_file("file_100MB.txt", 100)
```

```python
import matplotlib.pyplot as plt

# File sizes in MB
file_sizes = [200, 400, 600, 800, 1000]

# Time taken by each language (from the mini project sample table)
time_c =     [12, 20, 34, 45, 55]
time_cpp =   [15, 25, 36, 50, 60]
time_java =  [18, 30, 40, 55, 70]
time_r =     [20, 35, 45, 60, 80]
time_python = [25, 40, 53, 75, 100]
```

```python
plt.figure(figsize=(10,6))
plt.plot(file_sizes, time_c, label='C', marker='o')
plt.plot(file_sizes, time_cpp, label='C++', marker='s')
plt.plot(file_sizes, time_java, label='Java', marker='^')
plt.plot(file_sizes, time_r, label='R', marker='d')
plt.plot(file_sizes, time_python, label='Python', marker='x')

plt.title("Performance Comparison - Text File Conversion")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (sec)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
